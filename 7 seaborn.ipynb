## Contineous data plot
    distplot
    joinplot
    pairplot

```python
import seaborn as sns
```

```python
df = sns.load_dataset('tips') #Data about restaurant tip
```

```python
df.head()
# tip is TARGET VARIABLE | Dependent feature
```

```python
df.corr()
```

## Heatmap

```python
sns.heatmap(df.corr())
```

## join plot 
    Uni-variate | Bi-variate analysis

```python
sns.jointplot(x='tip', y='total_bill', data=df, kind='hex')
# explains density of data
# many people have given around Tip=2$, total_bill=10-20$
```

```python
sns.jointplot(x='tip', y='total_bill', data=df, kind='reg')
# reg = regression, it will draw probabitlity density functiion, regression line
```

## pair-plot
    mulri-variate analysis

```python
sns.pairplot(df)
```

```python
sns.pairplot(df, hue='sex')
# data is categorised using 'sex' feature
```

## Dist-plot
    helps to create histogram(Density | count)

```python
sns.distplot(df.tip)
# this gives you DENSITY %
```

```python
sns.distplot(df.tip, kde=False)
# This gives you COUNT
# KDE = Kernel density Estimation
```

## Categorical Data Plot
    box-plot
    violin-plot
    count-plot
    bar-plot

```python
# count-plot
# Categorical Features
df.select_dtypes(include='category').columns.tolist()
```

```python
sns.countplot('sex', data=df)
```

```python
sns.countplot(y='sex', data=df)
```

```python
import matplotlib.pyplot as plt
```

```python
cols = df.select_dtypes(include='category').columns.tolist()
i=1
for col in cols:
    plt.subplot(2,2,i)
    sns.countplot(col, data=df)
    i+=1
```

## bar-plot
    Accepts x & y values

```python
sns.barplot(x='sex', y='total_bill', data=df)
# by-default it is giving COUNTS
```

## Box-plot

```python
sns.boxplot('sex', 'total_bill', data=df)
```

```python
sns.boxplot('sex', 'total_bill', data=df, palette='rainbow')
```

```python
sns.boxplot('total_bill', 'sex', hue='smoker',data=df)
```

## Violin-plot
    helps us to see both Distribution of Data innterms of KDE and Box-plot

```python
sns.violinplot(x='total_bill', y='sex', data=df, palette='rainbow')
```
